```python
import os
import time
from pathlib import Path

# Step 1: Set your Student ID (must match the file created in Step 1)
student_id = "2025-4781" 

# Step 2: Define the path to the 'intro' file
file_path = Path.home() / "Documents" / "Activity_5_Files" / f"intro_{student_id}.txt"

# Step 3: Retrieve and display file metadata
if file_path.exists():
    # .stat() returns a collection of file attributes
    stat = file_path.stat()
    
    print(f"Student ID: {student_id}")
    print(f"File: {file_path.name}")
    
    # st_size tells you how much space the file takes on the disk
    print(f"Size: {stat.st_size} bytes")
    
    # st_mtime is a 'seconds since epoch' number, so we use time.ctime() 
    # to convert it into a human-readable date and time
    print(f"Last Modified: {time.ctime(stat.st_mtime)}")
else:
    print(f"File {file_path.name} not found for Student ID: {student_id}")
```

```python
from pathlib import Path

# Step 1: Set your Student ID (ensure consistency)
student_id = "2025-4781" 

# Step 2: Define the path to the 'lines' file
file_path = Path.home() / "Documents" / "Activity_5_Files" / f"lines_{student_id}.txt"

# Step 3: Safety Check - Create the file if it doesn't exist
if not file_path.exists():
    sample_lines = ["Line 1", "Line 2", "Line 3"]
    file_path.write_text("\n".join(sample_lines))
    print(f"Sample file created for Student ID: {student_id}")

# Step 4: Read the content and split into a list of strings
# .splitlines() is better than .split('\n') because it handles different OS line endings
lines = file_path.read_text().splitlines()

# Step 5: Process and rewrite the file
# We use "w" mode to overwrite the old text with the new formatted text
with file_path.open("w") as f:
    # enumerate(lines, 1) starts the counter at 1 instead of 0
    for i, line in enumerate(lines, 1):
        f.write(f"{i}: {line.upper()}\n")

# Step 6: Print confirmation
print(f"Lines formatted and updated in file: {file_path}")
```

```python
from pathlib import Path

# Step 1: Set your Student ID (ensure consistency)
student_id = "2025-4781" 

# Step 2: Define the path to the 'lines' file
file_path = Path.home() / "Documents" / "Activity_5_Files" / f"lines_{student_id}.txt"

# Step 3: Safety Check - Create the file if it doesn't exist
if not file_path.exists():
    sample_lines = ["Line 1", "Line 2", "Line 3"]
    file_path.write_text("\n".join(sample_lines))
    print(f"Sample file created for Student ID: {student_id}")

# Step 4: Read the content and split into a list of strings
lines = file_path.read_text().splitlines()

# Step 5: Reverse the list in-place
# This flips the order of the elements in the 'lines' list
lines.reverse()

# Step 6: Write the reversed lines back to the file
with file_path.open("w") as f:
    f.write("\n".join(lines))

# Step 7: Print confirmation
print(f"File lines reversed for Student ID: {student_id}")
```

```python
from pathlib import Path

# Step 1: Set your Student ID (ensure consistency)
student_id = "2025-4781" 

# Step 2: Define the directory and file paths
documents_path = Path.home() / "Documents" / "Activity_5_Files"
f1 = documents_path / f"intro_{student_id}.txt"
f2 = documents_path / f"lines_{student_id}.txt"
merged = documents_path / f"merged_{student_id}.txt"

# Step 3: Safety Check - Ensure both source files exist
if not f1.exists():
    f1.write_text(f"Welcome {student_id} to File Handling in Python!")
    print(f"Sample intro file created for Student ID: {student_id}")

if not f2.exists():
    f2.write_text("Line 1\nLine 2\nLine 3")
    print(f"Sample lines file created for Student ID: {student_id}")

# Step 4: Merge the contents of both files into a new file
# We open the 'merged' file in write mode ('w')
with merged.open("w") as mf:
    # Read and write content from the first file
    mf.write(f1.read_text())
    # Add a newline as a separator to prevent the texts from clumping together
    mf.write("\n")
    # Read and write content from the second file
    mf.write(f2.read_text())

# Step 5: Print confirmation
print(f"Files merged successfully for Student ID: {student_id}")
```
